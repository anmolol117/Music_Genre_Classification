```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import load_model
from tensorflow.keras.models import Sequential
from tensorflow.keras.initializers import HeNormal

```

```python
df = pd.read_csv("features_30_sec.csv")

df.head(5)
```

```python
X = df.drop(columns=["filename", "length", "label"])
y = df["label"]
```

```python
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)
```

```python
print("\nğŸ¯ Random Forest Results:")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf, target_names=label_encoder.classes_))
```

```python
cm = confusion_matrix(y_test, y_pred_rf)
plt.figure(figsize=(10, 6))
sns.heatmap(cm, annot=True, fmt="d", xticklabels=label_encoder.classes_,
            yticklabels=label_encoder.classes_, cmap="Blues")
plt.title("Random Forest Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.tight_layout()
plt.savefig("rf_confusion_matrix.png")
plt.close()
```

```python
y_train_cat = to_categorical(y_train, num_classes=10)
y_test_cat = to_categorical(y_test, num_classes=10)

he_init = HeNormal()

nn_model = Sequential([
    Dense(512, activation='relu', kernel_initializer=he_init, input_shape=(X_train.shape[1],)),
    BatchNormalization(),
    Dropout(0.4),

    Dense(256, activation='relu', kernel_initializer=he_init),
    BatchNormalization(),
    Dropout(0.3),

    Dense(128, activation='relu', kernel_initializer=he_init),
    BatchNormalization(),
    Dropout(0.2),

    Dense(64, activation='relu', kernel_initializer=he_init),
    BatchNormalization(),
    Dropout(0.2),

    Dense(10, activation='softmax')
])

nn_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

callbacks = [
    EarlyStopping(patience=10, restore_best_weights=True),
    ReduceLROnPlateau(patience=5, factor=0.5, verbose=1)
]

history = nn_model.fit(
    X_train, y_train_cat,
    epochs=100,
    batch_size=32,
    validation_data=(X_test, y_test_cat),
    callbacks=callbacks,
    verbose=1
)
```

```python
nn_loss, nn_accuracy = nn_model.evaluate(X_test, y_test_cat)
print(f"\nğŸ¤– Neural Network Accuracy: {nn_accuracy:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
import numpy as np

y_pred_nn_prob = nn_model.predict(X_test)
y_pred_nn = np.argmax(y_pred_nn_prob, axis=1)
y_true = np.argmax(y_test_cat, axis=1)  

cm_nn = confusion_matrix(y_true, y_pred_nn)

plt.figure(figsize=(10, 7))
sns.heatmap(cm_nn, annot=True, fmt="d", cmap="Blues",
            xticklabels=label_encoder.classes_,
            yticklabels=label_encoder.classes_)
plt.title("Neural Network Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.tight_layout()
plt.savefig("nn_confusion_matrix.png")
plt.close()
```

```python
joblib.dump(rf_model, "random_forest_genre.pkl")
joblib.dump(scaler, "scaler.pkl")
joblib.dump(label_encoder, "label_encoder.pkl")
nn_model.save("neural_net_genre.h5")
```
